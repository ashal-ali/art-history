```python
import pandas as pd
```

```python
modartists = pd.read_csv("/content/modernism_artists.csv")
modart = pd.read_csv("/content/modernism_artworks.csv", on_bad_lines='skip')
```

```python
modartists = modartists.rename(columns={"DisplayName": "Artist"})
```

```python
modart
```

```python
import plotly.express as px
```

```python
M_genders = ["(male)", "(male) (male)", "(male) (male)", "(male) (male) (male)"]
F_genders = ["(female)", "(female) (female)", "(female) (female) (female)"]

gender_mappings = {**{m: "Male" for m in M_genders}, **{f: "Female" for f in F_genders}}

modartists['Gender'] = modartists['Gender'].replace(gender_mappings)

print(modartists['Gender'].value_counts())
```

```python
modartists
```

```python
M_genders = ["(male)", "(male) (male)", "(male) (male)", "(male) (male) (male)"]
F_genders = ["(female)", "(female) (female)", "(female) (female) (female)"]

gender_mappings = {**{m: "Male" for m in M_genders}, **{f: "Female" for f in F_genders}}

modart['Gender'] = modart['Gender'].replace(gender_mappings)

print(modart['Gender'].value_counts())
```

```python
import pandas as pd
import plotly.express as px

# Convert "Date" to datetime, ignoring parsing errors
modart['Date'] = pd.to_datetime(modart['Date'], errors='coerce')

# Extract the year from the Date after conversion to datetime
modart['Year'] = modart['Date'].dt.year

# Filter the DataFrame to only include rows where "Gender" is "Male" or "Female"
# and where the "Year" is less than or equal to 2010
modart_filtered = modart[(modart['Gender'].isin(['Male', 'Female'])) & (modart['Year'] <= 2010)]

# Group the data by Year and Gender and count the artworks
gender_counts_over_time = modart_filtered.groupby(['Year', 'Gender']).size().reset_index(name='Artwork_Count')

# Create a line graph with the aggregated data
fig = px.line(gender_counts_over_time, x='Year', y='Artwork_Count', color='Gender',
              title='Gender Representation in Artworks Up to 2010',
              labels={'Year': 'Year of Creation', 'Artwork_Count': 'Number of Artworks', 'Gender': 'Artist Gender'})

# Show the plot
fig.show()
```
